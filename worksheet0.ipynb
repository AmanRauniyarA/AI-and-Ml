### Section 4.1: Exercise on Functions

```python
def convert_units():
    print('Choose conversion type:\n1. Length (m <-> ft)\n2. Weight (kg <-> lbs)\n3. Volume (L <-> gal)')
    choice = input('Enter 1, 2, or 3: ')
    try:
        val = float(input('Enter the value to convert: '))
        if choice == '1':
            direction = input('Enter A for m to ft, B for ft to m: ').upper()
            result = val * 3.28084 if direction == 'A' else val / 3.28084
            print(f'Result: {result} ft' if direction == 'A' else f'Result: {result} m')
        elif choice == '2':
            direction = input('Enter A for kg to lbs, B for lbs to kg: ').upper()
            result = val * 2.20462 if direction == 'A' else val / 2.20462
            print(f'Result: {result} lbs' if direction == 'A' else f'Result: {result} kg')
        elif choice == '3':
            direction = input('Enter A for L to gal, B for gal to L: ').upper()
            result = val * 0.264172 if direction == 'A' else val / 0.264172
            print(f'Result: {result} gal' if direction == 'A' else f'Result: {result} L')
        else:
            print('Invalid conversion type.')
            return None
        return result
    except ValueError:
        print('Error: Invalid input. Please enter numeric values.')
        return None

def list_operations():
    print('Choose operation:\n1. Sum\n2. Average\n3. Maximum\n4. Minimum')
    choice = input('Enter 1, 2, 3, or 4: ')
    try:
        nums_str = input('Enter numbers separated by spaces: ')
        nums = [float(x) for x in nums_str.split()]
        if not nums:
            raise ValueError('List cannot be empty.')
        if choice == '1':
            res = sum(nums)
        elif choice == '2':
            res = sum(nums) / len(nums)
        elif choice == '3':
            res = max(nums)
        elif choice == '4':
            res = min(nums)
        else:
            print('Invalid operation choice.')
            return None
        print(f'Result: {res}')
        return res
    except ValueError as e:
        print(f'Error: {e}. Please ensure you enter valid numeric values.')
        return None
        # This will actually run the functions
convert_units()
list_operations()
```

### Section 4.2: Exercise on List Manipulation

```python
def extract_every_other(lst):
    return lst[::2]

def get_sublist(lst, start, end):
    return lst[start:end+1]

def reverse_list(lst):
    return lst[::-1]

def remove_first_last(lst):
    return lst[1:-1]

def get_first_n(lst, n):
    return lst[:n]

def get_last_n(lst, n):
    return lst[-n:]

def reverse_skip(lst):
    return lst[-2::-2]

# Testing the functions
test_list = [1, 2, 3, 4, 5, 6]
print('1. Extract every other:', extract_every_other(test_list))
print('2. Sublist (2 to 4):', get_sublist(test_list, 2, 4))
print('3. Reverse:', reverse_list(test_list))
print('4. Remove first and last:', remove_first_last(test_list))
print('5. First 3 elements:', get_first_n(test_list, 3))
print('6. Last 2 elements:', get_last_n(test_list, 2))
print('7. Reverse skip:', reverse_skip(test_list))
```

### Section 4.3: Exercise on Nested List

```python
def deep_flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list

def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]
    return element

def sum_nested(lst):
    return sum(deep_flatten(lst))

def remove_element(lst, elem):
    modified_list = []
    for item in lst:
        if isinstance(item, list):
            modified_list.append(remove_element(item, elem))
        elif item != elem:
            modified_list.append(item)
    return modified_list

def find_max(lst):
    return max(deep_flatten(lst))

def count_occurrences(lst, elem):
    return deep_flatten(lst).count(elem)

def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat) if flat else 0

flatten = deep_flatten
# 1. Flatten a Nested List
test_list_1 = [[1, 2], [3, 4], [5]]
print("1. Flatten:", flatten(test_list_1))

# 2. Accessing Nested List Elements
test_list_2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("2. Access Element [1, 2]:", access_nested_element(test_list_2, [1, 2]))

# 3. Sum of All Elements
test_list_3 = [[1, 2], [3, [4, 5]], 6]
print("3. Sum Nested:", sum_nested(test_list_3))

# 4. Remove Specific Element
test_list_4 = [[1, 2], [3, 2], [4, 5]]
print("4. Remove '2':", remove_element(test_list_4, 2))

# 5. Find the Maximum Element
print("5. Find Max:", find_max(test_list_3))

# 6. Count Occurrences
test_list_6 = [[1, 2], [2, 3], [2, 4]]
print("6. Count '2's:", count_occurrences(test_list_6, 2))

# 7. Flatten a List of Lists of Lists
test_list_7 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print("7. Deep Flatten:", deep_flatten(test_list_7))

# 8. Nested List Average
test_list_8 = [[1, 2], [3, 4], [5, 6]]
print("8. Average:", average_nested(test_list_8))
```
